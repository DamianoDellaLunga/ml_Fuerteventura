## Line plot

2. Create a 1d-array of 30 data points from range 10 to range 40 and use it like the **x** variable. Create the **y** variable by multiply **x * 2**.
 * Plot the data in a line plot

```python
# Creating x and y
x=np.arange(10,40)
y=x*2
```

```python
# Set the figure size
plt.figure(figsize=(8,5))
plt.plot(x,y,'g')

# Plotting the data
plt.title('Boh')


# Setting the title



# Setting the axis names
plt.xlabel('number of bananas')
plt.ylabel('number of monkeys')

# show
plt.show()
```

### Multiple lines on the same plot

5. Use the data belowe. Plot them on the same plot, use color red for **y1** and green for **y2**

```python
# data
x = np.arange(1,11)
y1 = x ** 2
y2 = x ** 3
```

```python
# figure size
plt.figure(figsize=(8,5))

# plot function
plt.plot(x,y1, 'r*', label='Google')
plt.plot(x,y2, 'ko', label='Vatican')
# title
plt.title('Happy endings')
plt.legend()
# axis
plt.xlabel('Number of employers')
plt.ylabel('Number of HJ')
# show
plt.show()
```

6. Use the data below and plot the in a barchart
