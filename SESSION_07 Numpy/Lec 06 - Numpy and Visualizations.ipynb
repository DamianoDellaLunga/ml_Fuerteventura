**matplotlib.pyplot** is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
# import numpy
import numpy as np

# here we import a specific package
import matplotlib.pyplot as plt

# this code is needed to visualize the pots
%matplotlib inline
```

## First example

#### Line plot

```python
# Creating the array
x = np.arange(2, 21)
y = 2 * x + 5
```

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'b', label='Data') # 'b'=blue color

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.legend()

plt.show()
```

#### Variation of line plot 

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'bo', label='Data') #added new parameter

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.legend()
plt.show()
```

## Second Example

#### Exponential plot

```python
# Calculate exponential for all elements in the array.
x = np.arange(1,11)
y = np.exp(x)
```

```python
# Make a new plot
plt.figure(figsize=(8,4))

# plot function
plt.plot(x, y, 'g', label='my data')

# setting the title
plt.title('Exponential Plot')

# setting axis
plt.xlabel('Samples')
plt.ylabel('Exponential')
plt.legend()

# show
plt.show()
```

### Multiple lines on the same plot

```python
# data
x = np.arange(1,11)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
```

```python
plt.figure(figsize=(8,4))

# dashes, squares, triangles
plt.plot(x, y1, 'g--', label='Squared') 
plt.plot(x, y2, 'rs', label='Cubic')
plt.plot(x, y3, 'b^', label='Forth')

plt.title('Multiple Plot')
plt.xlabel('Samples')
plt.ylabel('Values')
plt.grid()
plt.legend()
plt.show()
```

### Variation of the exponential plot (bar chart) 

```python
# data
x = np.arange(10,21)
y = x ** 2
```

```python
# Make a new plot
plt.figure(figsize=(8,4))

# plot function
plt.bar(x, y)

# title and axis
plt.title('Bar Plot')
plt.xlabel('Samples')
plt.ylabel('Squared')

#show
plt.show()
```

### Scatterplot

```python
# data
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.figure(figsize=(10,5))
plt.scatter(x,y, c='r')

plt.show()
```

## Multiple Plots

 * The **fig** instance represents an empty canvas
 * The **axes** represents an array of axes where to plot our data


```python
# The fig instance represents an empty canvas
# Axes is an array of axes to plot on

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,5))

# Plot 1
axes[0].plot(x, y**2, 'r--')
axes[0].set_title('Title 1')
axes[0].set_ylabel('Value ** 3')

# Plot 2
axes[1].plot(x, y, 'g^')
axes[1].set_title('Title 2')
axes[1].set_ylabel('Value Linear')

plt.show()
```
