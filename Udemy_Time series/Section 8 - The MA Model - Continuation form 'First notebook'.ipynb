## Importing the relevant packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.arima_model import ARMA
from scipy.stats.distributions import chi2 
from math import sqrt
import seaborn as sns
sns.set()
```

## Importing the Data and Pre-processing 

```python
raw_csv_data = pd.read_csv("Index2018.csv") 
df_comp=raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.set_index("date", inplace=True)
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

```python
df_comp['market_value']=df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## The LLR Test

```python
def LLR_test(mod_1, mod_2, DF = 1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))    
    p = chi2.sf(LR, DF).round(3)
    return p
```

## Creating Returns

```python
df['returns'] = df.market_value.pct_change(1).mul(100)
```

## ACF for Returns

```python
sgt.plot_acf(df.returns[1:], lags=40, zero=False)
plt.title("ACF returns", size=24)
plt.show()
```

## MA(1) for Returns

```python
model_ret_ma_1=ARMA(df.returns[1:], order=(0,1))
results_ret_ma_1=model_ret_ma_1.fit()
results_ret_ma_1.summary()
```

## Higher-Lag MA Models for Returns

```python
model_ret_ma_2=ARMA(df.returns[1:], order=(0,2))
results_ret_ma_2=model_ret_ma_2.fit()
print(results_ret_ma_2.summary())
print("\nLLR test p-value=" + str(LLR_test(model_ret_ma_1, model_ret_ma_2)))
```

```python
model_ret_ma_3=ARMA(df.returns[1:], order=(0,3))
results_ret_ma_3=model_ret_ma_3.fit()
print(results_ret_ma_3.summary())
print("\nLLR test p-value=" + str(LLR_test(model_ret_ma_2, model_ret_ma_3)))
```

```python
model_ret_ma_6=ARMA(df.returns[1:], order=(0,6))
results_ret_ma_6=model_ret_ma_6.fit()
print(results_ret_ma_6.summary())
```

```python
model_ret_ma_7=ARMA(df.returns[1:], order=(0,7))
results_ret_ma_7=model_ret_ma_7.fit()
print(results_ret_ma_7.summary())
print("\nLLR test p-value=" + str(LLR_test(model_ret_ma_6, model_ret_ma_7)))
```

```python
model_ret_ma_8=ARMA(df.returns[1:], order=(0,8))
results_ret_ma_8=model_ret_ma_8.fit()
print(results_ret_ma_8.summary())
print("\nLLR test p-value=" + str(LLR_test(model_ret_ma_7, model_ret_ma_8)))
```

```python
LLR_test(model_ret_ma_6, model_ret_ma_8, DF=2)
```

```python
# The low p-values suggest that MA(8) is superior to MA(6)!
```

## Residuals for Returns

```python
df['res_ret_ma_8']=results_ret_ma_8.resid[1:]
```

```python
print("The mean of residuals is" + str(round(df.res_ret_ma_8.mean(),3)) + "\nThe variance of the residuals is" + str(round(df.res_ret_ma_8.var(),3)))
```

```python
round(sqrt(df.res_ret_ma_8.var()),3)
```

```python
df.res_ret_ma_8[1:].plot(figsize=(20,5))
plt.title("Residuals of Returns", size=24)
plt.show()
```

```python
sts.adfuller(df.res_ret_ma_8[2:])
```

```python
## the data of the residulas is stationary but it may not be white noise
```

```python
sgt.plot_acf(df.res_ret_ma_8[2:], lags=40, zero=False)
plt.title("ACF Residuals of Returns", size=24)
plt.show()
```

###### If the data was alike White noise, all the coefficient should be non-significant. Here is similar, whit the first 17 coefficient that are non-significant

## Normalized Returns

```python
bench_ret=df.returns.iloc[1]
df['norm_ret']=df.returns.div(bench_ret).mul(100)
```

```python
sgt.plot_acf(df.norm_ret[1:], lags=40, zero=False)
plt.title("ACF for Normalized returns", size=24)
plt.show()
```

```python
model_norm_ret_ma_8=ARMA(df.returns[1:], order=(0,8))
results_norm_ret_ma_8=model_norm_ret_ma_8.fit()
results_norm_ret_ma_8.summary()
```

```python
df['res_norm_ret_ma_8']=results_ret_ma_8.resid[1:]
df.res_norm_ret_ma_8[1:].plot(figsize=(20,5))
plt.title("Residuals of Returns", size=24)
plt.show()
```

```python
sgt.plot_acf(df.res_norm_ret_ma_8[2:], lags=40, zero=False)
plt.title("ACF for Residuals of Normalized returns", size=24)
plt.show()
```

##### The data resemble white noise, so our model is likely correct
