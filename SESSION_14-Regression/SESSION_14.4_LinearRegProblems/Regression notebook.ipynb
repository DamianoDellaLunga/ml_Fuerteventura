```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#data generation for linear regression
```

```python
area_house=np.random.rand(500)*10
```

```python
area_house
```

```python
len(area_house)
```

```python
prices=7*area_house+3
```

```python
plt.scatter(area_house, prices)
```

```python
#we modify the dataset to be more realistic
prices=prices+np.random.randn(500)*5
```

```python
plt.scatter(area_house, prices)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
my_linear_model=LinearRegression()
```

```python
my_linear_model.fit(area_house, prices) #x and y
```

```python
area_house1=area_house.reshape(500,1) # you can put -1 instead of 500, that means we ask python to find itself the number of rows that it needs
```

```python
area_house
```

```python
area_house1
```

```python
my_linear_model.fit(area_house1, prices) #x and y again
```

```python
my_linear_model.coef_ #this get the slope
```

```python
my_linear_model.get_params
```

```python
my_linear_model.intercept_
```

```python
#we want to plot the model now
plt.scatter(area_house1,prices)
plt.grid()
```

```python
start=my_linear_model.predict(np.array([[0]]))
end=my_linear_model.predict(np.array([[10]]))
```

```python
list(end)[0]
```

```python
plt.scatter(area_house1,prices)
plt.plot([0,10],[list(start)[0], list(end)[0]], 'r', linewidth=3)
#a simple x-y plot with a list of number 0-10 as x and a straight line with the first value of 'start'
#and the first value of 'end'.
```
